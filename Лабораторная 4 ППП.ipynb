```python
df['Victim Sex'] = df['Victim Sex'].replace({"M": 1, "F": -1})
df = df[~df['Victim Sex'].isin(['H', 'X', np.nan])]
df["Victim Sex"].unique()
for v_d in df["Victim Descent"].unique():
    subset = df[df["Victim Descent"] == v_d]
    print(f"Люди происхожения {v_d}")
    display(subset[['Victim Sex', "Victim Age"]].corr())
```

```python
df['Victim Age'].hist()
```

чаще всего жертвами становятся люди от 20 до 40

```python
density = sp.stats.gaussian_kde(df['Victim Age'].dropna())
 
x_vals = np.linspace(0,100, 1000) # Specifying the limits of our data
density.covariance_factor = lambda : .5 #Smoothing parameter
 
density._compute_covariance()
plt.plot(x_vals,density(x_vals))
plt.show()
```

В целом на графике так и показано, только с более точными числами - вероятность более 1%(т.е. большинство) - люди в возрасте от 15 до 60 лет

# Часть 2

```python
import cv2
```

```python
base_img = cv2.imread("4 R2D2.jpg")
plt.imshow(base_img)
cv2.imshow("", base_img)
cv2.waitKey(0)
height, width, channels = base_img.shape
print(height, width, channels)
```

Считал картинку и вывел ее размеры(и саму картинку)

```python
half_width = width // 2
leftrighted_img = base_img.copy()
for y in range(height):
    # Swap left and right halves of each row
    left_half = base_img[y, :half_width].copy()
    right_half = base_img[y, half_width:].copy()
    leftrighted_img[y, :half_width] = right_half
    leftrighted_img[y, half_width:] = left_half[::-1]
plt.imshow(leftrighted_img)
cv2.waitKey(0)
```

поменял стороны местами и инвертировал картинку справа

```python
inverted_img = leftrighted_img.copy()
for y in range(half_width, width):
    for x in range(height):
        inverted_img[x, y,0] = 255 - leftrighted_img[x, y,0]
        inverted_img[x, y,1] = 255 - leftrighted_img[x, y,1]
        inverted_img[x, y,2] = 255 - leftrighted_img[x, y,2]
plt.imshow( inverted_img)
cv2.waitKey(0)
```

инвертировал правую половину

```python
cv2.imwrite("4 R2D2_new.jpg", inverted_img)
```

записал картинку в файл
