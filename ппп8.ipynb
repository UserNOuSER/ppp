## Задание 5. Реализуйте интерфейс для одного из заданий

Окно должно содержать кнопку для открытия выбранного файла. \
Содержимое файла после открытия отображается в двух Label или Text (массивы x, y). \
Три окна Text позволяют задать начальное значение, шаг, конечное значение для расчёта промежуточных значений (необходима проверка корректности введённых данных). \
Результаты расчёта выводятся на графике в новом окне (plt.show): исходные точки отображаются чёрными квадратами, а новые – зелёными квадратами, соединёнными прямыми линиями.

```python
from tkinter import filedialog
import matplotlib.pyplot as plt
from functools import partial
import numpy as np
from scipy.interpolate import interp1d

def openfile():
    path = tk.filedialog.askopenfilename(filetypes=[("Текстовые файлы", "*.txt")])
    if path:
        print("Файл выбран:", path)
        interpol(path)

def interpol(path):
    with open(path, "r") as file:
        str_x = file.readline()
        str_y = file.readline()
        x = list(map(int, str_x.split()))
        y = list(map(int, str_y.split()))
    data_window = tk.Tk()
    data_window.title("Вывод данных с файла")
    
    tk.Label(master = data_window, text = "x =" + str_x).grid(row = 0, column = 0)
    tk.Label(master = data_window, text = "y =" + str_y).grid(row = 1, column = 0)

    tk.Label(master = data_window, text = "Начальное значение").grid (row = 0, column = 1)
    tk.Label(master = data_window, text = "Шаг").grid (row = 1, column = 1)
    tk.Label(master = data_window, text = "Конечное значение").grid (row = 2, column = 1)

    vcmd = data_window.register(is_valid)
    
    ent_begin = tk.Entry(master = data_window, validate="key", validatecommand=(vcmd, "%P"))
    ent_diff = tk.Entry(master = data_window, validate="key", validatecommand=(vcmd, "%P"))
    ent_end = tk.Entry(master = data_window, validate="key", validatecommand=(vcmd, "%P"))

    btn_draw = tk.Button(master = data_window, text = "Нарисовать график", command = partial(draw_plot, ent_begin,  ent_diff, ent_end, x, y))

    ent_begin.grid (row = 0, column = 2)
    ent_diff.grid (row = 1, column = 2)
    ent_end.grid (row = 2, column = 2)
    btn_draw .grid (row = 3, column = 2)
    data_window.mainloop()

def is_valid(newval):
    if newval == "":
        return True
    try:
        float(newval)
        return True
    except ValueError:
        return False
    
def draw_plot(e_begin, e_diff, e_end, x, y):
    begin, diff, end = e_begin.get(),  e_diff.get(), e_end.get()
    if begin!= "" and diff!= "" and end!= "":
        print("Числа введены корректно")
        new_x = np.arange(float(begin), float(end), float(diff))
        plt.plot(x, y, 'ks') #Базовый график
        plt.plot(new_x, interp1d(x, y, kind='quadratic')(new_x), 'gs-') #Интерполированный квадратичным
        plt.show()
    else:
        print("Ошибка: введено не число!", begin, diff, end)
        return 
    
file_window = tk.Tk()
file_window.title("Открытие файла")
btn_openfile = tk.Button(master = file_window, text = "Открыть файл", command = openfile)
btn_openfile.pack() 
file_window.mainloop()

```
