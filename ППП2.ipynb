## Задание 4 ##

 Создайте функцию: а) с помощью ключевого слова def, б) с помощью ключевого слова lambda (lambda-функция), - для расчёта функции вида y=f(x) в заданной точке. Постройте график этой функции в диапазоне от 0 до x0 с шагом 0.1 с помощью библиотеки matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

def chetofunc(x):
    return cmath.exp(cmath.sqrt(cmath.sin(x)) - 1.5)
x = 3.64
y = lambda x: cmath.exp(cmath.sqrt(cmath.sin(x)) - 1.5)
print(y(x))
print(chetofunc(x))
```

```python
x = list(np.arange(0,3.64, 0.1))
y_arr = [y(i) for i in x]
plt.figure()
plt.plot(x, y_arr)
plt.xlabel('x'), plt.ylabel('y')
plt.show()
```

## Задание 9 ##

Подсчёт количества элементов меньших и больших некоторого числа в матрице произвольного размера. Функция на вход получает матрицу, и индексы строки, столбца элемента, а возвращает список из двух элементов, где первый - количество элементов меньше указанного, а второй - больше.

```python
from random import randint
width = randint(2,10)
height = randint(2,10)
matr = [[randint(0,10) for _ in range(width)] for ___ in range(height)]
print(*matr, sep = "\n")
pos_y, pos_x = map(int, input("Введите координаты \n").split())
bigger = []
smaller = []
number = matr[pos_x][pos_y]
print("Выбранное вами число -", number)
for i in range(height):
    for j in range(width):
        if matr[i][j] > number:
            bigger +=[matr[i][j]]
        elif matr[i][j] < number:
            smaller +=[matr[i][j]]
print(*bigger)
print(*smaller)
```
